```python
import os
import pandas as pd
```

```python
# Path to the data folder
data_folder = '../data/'

# Iterate through all files in the folder
for file_name in os.listdir(data_folder):
    file_path = os.path.join(data_folder, file_name)
    
    # Check if the file is a CSV file
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_path)
        print(f"File: {file_name} -> Rows: {df.shape[0]}, Columns: {df.shape[1]}")

        # count total rows with duplicate values based on product_name only
        product_name_counts = df['product_name'].value_counts()
        print("Product Name Counts: ", product_name_counts)
        

print("Total Files: ", len(os.listdir(data_folder)))
```

```python
tr = 0

# Iterate through all files in the folder
for file_name in os.listdir(data_folder):
    file_path = os.path.join(data_folder, file_name)
    
    # Check if the file is a CSV file
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_path)

        print(f"Total Rows of {file_name} is {df.shape[0]} ")
        df = df.drop_duplicates(subset='product_name', keep='first')
        print(f"DataFrame after removing duplicates: {df.shape[0]} rows")
        tr += df.shape[0]


print("Total Unique Rows: ", tr)

```

```python
# Initialize an empty list to store DataFrames
dataframes = []

# Iterate through all files in the folder
for file_name in os.listdir(data_folder):
    file_path = os.path.join(data_folder, file_name)
    
    # Check if the file is a CSV file
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dataframes, ignore_index=True)

print(f"Merged DataFrame has {merged_df.shape[0]} rows and {merged_df.shape[1]} columns.")
```

```python
# Remove duplicate rows based on product_name column
unique_df = merged_df.drop_duplicates(subset=['product_name'])

print(f"Unique DataFrame has {unique_df.shape[0]} rows and {unique_df.shape[1]} columns.")
```

```python
unique_categories = unique_df['category'].value_counts()
# print(f"Unique Categories: {unique_categories}")

# print the unique categories name
print(unique_categories.index.tolist())
```

```python
unique_df.to_csv('merged_10K.csv', index=False)
```
